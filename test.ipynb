```python
import yt_dlp
import json

def get_transcript_with_ytdlp(video_url):
    """Get transcript using yt-dlp (supports more videos)"""
    ydl_opts = {
        'skip_download': True,
        'writesubtitles': True,
        'writeautomaticsub': True,
        'subtitleslangs': ['en'],
        'subtitlesformat': 'json3',
        'quiet': True,
    }
    
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video_url, download=False)
            
            if 'en' in info.get('subtitles', {}):
                # Get manual subtitles
                subtitle_url = info['subtitles']['en'][0]['url']
            elif 'en' in info.get('automatic_captions', {}):
                # Get automatic captions
                subtitle_url = info['automatic_captions']['en'][0]['url']
            else:
                return "No English subtitles available"
            
            # Download and parse subtitle JSON
            import urllib.request
            response = urllib.request.urlopen(subtitle_url)
            subtitle_data = json.loads(response.read())
            
            # Extract text
            full_text = ""
            for event in subtitle_data.get('events', []):
                if 'segs' in event:
                    for seg in event['segs']:
                        full_text += seg.get('utf8', '') + " "
            
            return full_text.strip()
            
    except Exception as e:
        return f"Error: {str(e)}"

# Usage
# pip install yt-dlp
video_url = "https://youtu.be/Y5f7mx6ES6Q?si=Yguwtv85oLBW7Q5B"
transcript = get_transcript_with_ytdlp(video_url)
print(transcript)
```
